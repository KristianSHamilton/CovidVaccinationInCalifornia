```python
#Code for cleaning the administered date in Tikaram's notebook
#this will take all individual days in the column 'administered_date'
#and add a column showing 'admistered_month', which we can later use to merge
#our dataframes (yours on vaccinations, and mine on cases) by month
import os
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import requests
import gmaps

from sodapy import Socrata
low_memory = False
from pandas.api.types import CategoricalDtype
```

```python
CHHS_Path = "./130d7ba2-b6eb-438d-a412-741bde207e1c (1).csv"
CHHS_File = pd.read_csv(CHHS_Path)
chhs_df = pd.DataFrame(CHHS_File)
chhs_df
```

```python
#take data from 'adminstered_date' and copy it into a separate list
date_list = chhs_df['administered_date'].tolist() 
#check to make sure the list and pd column have the same number of rows
print(len(date_list))

#make blank list to hold the month data only, iterate through the dates,
#and return only the year and month
month_list = []
for day in date_list:
    month_list.append(day[0:7])

```

```python
#convert list to dataframe, and join the dataframes
month_list_df = pd.DataFrame(month_list, columns=['Month Administered'])
chhs_df = chhs_df.join(month_list_df)
chhs_df
```

```python
#extract only the data for dates in 2021, except for August
chhs_df_2021 = chhs_df.loc[(chhs_df['Month Administered'] == '2021-01') | (chhs_df['Month Administered'] == '2021-02') | 
                              (chhs_df['Month Administered'] == '2021-03') | (chhs_df['Month Administered'] == '2021-04')  
                            | (chhs_df['Month Administered'] == '2021-05') | (chhs_df['Month Administered'] == '2021-06') 
                            | (chhs_df['Month Administered'] == '2021-07')] 
chhs_df_2021
```
